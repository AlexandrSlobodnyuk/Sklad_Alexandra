```python
from numbers import Number
import itertools
from math import sqrt

class ComplexnumbersError(ValueError):
    pass
class ComplexNumb:
    def __init__(self, arg=0):
        if isinstance(arg, Number):
            self.coefficients = [arg]
        elif isinstance(arg, list):
            self.coefficients = arg.copy()
        elif isinstance(arg, ComplexNumb):
            self.coefficients = arg.coefficients.copy()
        else:
            raise ComplexnumbersError("You are trying to create сomplex from" + repr(arg))
        if len(self.coefficients) > 2:
            for _ in range(len(self.coefficients) - 2): 
                del self.coefficients[-1]
        if len(self.coefficients) == 1:
            self.coefficients += [0]
        if len(self.coefficients) == 0:
            self.coefficients += 2 * [0]
            

    def __str__(self):
        return (" + ".join([
            str(c) + ("*" + i if i != '' else '')
            for c, i in list(zip(self.coefficients, ['','j'])) if c != 0
            ])) if len(self.coefficients) and sum(self.coefficients) else '0'


    def __eq__(self, other):
        if isinstance(other, Number):
            other = Complex(other)
        
        if isinstance(other, Complex):
            return self.coefficients == other.coefficients
        else:
            raise ComplexnumbersError("Can't say if complex is equal to " + str(type(other)))

    def __add__(self, other):
        if isinstance(other, Number):
            other = ComplexNumb(other)
        
        selfcoefs = self.coefficients.copy()
        othercoefs = other.coefficients.copy()
        
        return ComplexNumb([
            sce + oce for sce, oce in zip(selfcoefs, othercoefs)
        ])


    def __radd__(self, other):
        return self.__add__(other)


    def __neg__(self):
        return Complex([-c for c in self.coefficients])


    def __sub__(self, other):
        if isinstance(other, Number):
            other = ComplexNumb(other)

        return self.__add__(other.__neg__())


    def __rsub__(self, other):
        return self.__neg__().__add__(other)

    def __mul__(self, other):
        if isinstance(other, Number):
            other = ComplexNumb(other)
        if isinstance(self, Number):
            self = ComplexNumb(self)
        s = self.coefficients.copy()    
        o = other.coefficients.copy()
        return ComplexNumb([(s[1] * o[0] + s[0] * o[1]), 
                            (s[0] * o[0] - o[1] * s[1])])

    def __invert__(self):
        #Комплексное сопряжение
        return ComplexNumb([self.coefficients[0]] + [-1 * self.coefficients[1]])


    def __abs__(self):
        #Модуль комплексного числа
        return sum(i**2 for i in self.coefficients)


    def __truediv__(self, other):
        #Деление
        if isinstance(self, Number):
            self = ComplexNumb(self)
        if isinstance(other, Number):
            other = ComplexNumb(other)

        if other.coefficients == []:
            raise ComplexnumbersError("Dividing by zero")
        else:
            return self.__mul__(ComplexNumb([i/(other.__abs__()) for i in other.__invert__().coefficients]))

     
    def __complex__(self):
        return complex(self.coefficients[0], self.coefficients[1])

    def __float__(self):
        if self.coefficients[1] == 0:
            return float(self.coefficients[0])
        else:
            raise ComplexnumbersError("Converting error")
        

    def __int__(self) -> int:
        if self.coefficients[1] == 0:
            return int(self.coefficients[0])
        else:
            raise ComplexnumbersError(" Converting error") 
    

                       
                    
```
